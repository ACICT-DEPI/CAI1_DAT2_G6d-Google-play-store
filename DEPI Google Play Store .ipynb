###  Fix Reviews

```python
for review in df['Reviews'].unique():
    print(review)
```

```python
# Changing the data type of 'Reviews' and 'Installs' from float to int
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
df.info()
```

### 4- Fix size

```python
for s in df['Size'].unique():
    print(s)
```

```python
def fix_size(s):
    try:
        if s.endswith('M'):
            return float(s[:-1])
        elif s.endswith('k'):
            return float(s[:-1]) / 1024
        elif s == '1,000+':
            return 1.0
        else:
            return np.nan
    except:
        return np.nan
```

```python
df['Size'] = df['Size'].apply(fix_size)
df.info()
```

```python
df.head()
```

### Fix Installs

```python
for installs in df['Installs'].unique():
    print(installs)
```

```python
#Removing the + symbol
df['Installs'] = df.Installs.apply(lambda x: x.strip('+'))
df['Installs'].info()
```

```python
# Removing the ; symbol
df['Installs'] = df.Installs.apply(lambda x: x.replace(',' , ''))
df['Installs']
```

```python
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
df.info()
```

### 7- Fix price

```python
for p in df['Price'].unique():
    print(p)
```

```python
def fix_price(p):
    try:
        if p.startswith('$'):
            return float(p[1:])
        else:
            return 0.0
    except:
        return np.nan
```

```python
df['Price'] = df['Price'].apply(fix_price)
df.info()
```

```python
df.describe()
```

### Filling the missing value

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer()
```

```python
df['Rating'] = imputer.fit_transform(df[['Rating']])
df['Size'] = imputer.fit_transform(df[['Size']])
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Current Ver'].unique()
```

```python
df['Android Ver'].unique()
```

```python
def clean_android_ver(version):
    if pd.isna(version):  # Check if the value is NaN
        return None
    elif 'Varies with device' in version:
        return None  # Return None for 'Varies with device'
    else:
        return version.split(' and')[0] 
```

```python
# Apply the cleaning function to 'Android Ver' column
df['Cleaned Android Ver'] = df['Android Ver'].apply(clean_android_ver)
```

```python
df.info()
```

```python
# Drop rows with NaN in 'Cleaned Android Ver' (if you want to remove rows with invalid version info)
df_cleaned = df.dropna(subset=['Cleaned Android Ver'])

# Display the cleaned DataFrame
df_cleaned
```

```python
# Dropping columns irrelevant to analysis
df_cleaned.drop(['Current Ver', 'Android Ver'], axis=1, inplace= True)
```

```python
df_cleaned.info()
```

###  duplicate values in 'App'

```python
# Remove duplicates by keeping the row with the maximum "Reviews" for each "App"
df_cleaned = df.loc[df.groupby('App')['Reviews'].idxmax()]

# Save the cleaned data to a new file
df_cleaned.to_csv('data/cleaned_playstore2.csv', index=False)

# Display the first few rows of the cleaned data
print(df_cleaned.head())
```

```python
# Saving the cleaned Dataset to a csv for further Analysis and Visualisation in SQL Server and PowerBI
df_cleaned.to_csv('data/cleaned_playstore.csv')
```
